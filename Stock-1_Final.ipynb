A skewness of +1 indicates a positive skewness, which means that the distribution of last prices is skewed to the right.

In a positively skewed distribution, the tail on the right side is longer and fatter than the left side. This suggests that there may be some higher-than-average last prices in the data. It also suggests that there may be a concentration of relatively high last prices in the dataset. 

Higher-than-average last prices could reflect positive market sentiment or strong demand for the stock.

Now, the combination of positive skewness and positive kurtosis suggests that while there is a tendency for occasional higher prices, the distribution overall has relatively lighter tails, meaning extreme values are not as prevalent as they would be in a distribution with heavier tails.

***Justification***

Due to the less stability of the comapany, There is always fear of investing in the company even though there are few evidences of stability in stock prices. We need an advantage to this disadvantage. So, we looked at the skewness and kurtosis of the last price. 

The last price reflects the most recent transaction in the market. It provides real-time information about the current value of a stock, allowing us to gauge the latest market sentiment. 

Skewness measures the asymmetry of a probability distribution. A negative skewness indicates a distribution that is skewed to the left (tail to the left), while positive skewness indicates a distribution that is skewed to the right (tail to the right). So, Skewness helps us to assess whether the distribution of last prices is symmetric or pushed in one direction. A skewed distribution might suggest a prevailing trend in the stock prices. Negative skewness might be associated with a distribution where large negative returns are more frequent, while positive skewness might indicate a distribution with more frequent positive returns.

Kurtosis measures the "tailedness" of a probability distribution. It assesses whether the tails of the distribution are fatter or thinner than those of a normal distribution. High kurtosis i.e. greater than 1 may indicate the presence of fat tails in the distribution, suggesting a higher probability of extreme events or outliers in stock returns. 

Understanding the skewness and kurtosis of the last prices can provide valuable information about the distribution of returns and potential risks.

```python
df['Date'] = pd.to_datetime(df['Date'])


df['month'] = df['Date'].dt.month
df['year'] = df['Date'].dt.year


monthly_stats = df.groupby(['year', 'month']).agg({
    'Open': 'first',
    'Close': 'last',
    'High': 'max',
    'Low': 'min',
    'Last': 'last',
    'Turnover': 'sum',
    'VWAP': 'mean'
}).reset_index()


plt.figure(figsize=(12, 6))

plt.plot(monthly_stats['year'].astype(str) + '-' + monthly_stats['month'].astype(str), monthly_stats['Close'], marker='o', linestyle='-', color='b')
plt.title('Monthly Closing Prices (2020-2021)')
plt.xlabel('Month')
plt.ylabel('Closing Price')
plt.xticks([])
plt.show()


monthly_stats['monthly_return'] = monthly_stats['Close'].pct_change() * 100


print(monthly_stats[['year', 'month', 'monthly_return']])
```

***Conclusion***

There has been an gradual increase in the closing price of the comapny in the start due its financial growth and then there is sudden increase which might be due to high demand of stocks. Later there is a drop after a drastic increase which might be the usual trends of stock market. The company has later picked up its prices. Overall, there has been stability in the closing prices with less fluctuations. Some dots are concentrated at a specific levels. It could indicate certain market events, such as product launches, or economic reports, can lead to significant price movements or some companies may experience seasonal trends that impact their stock prices. As this is a retail companies, it may see increased activity around the holiday season.

```python
le = LabelEncoder()
df['encoded_year'] = le.fit_transform(df['year'])

yearly_open_prices = df.groupby('encoded_year')['VWAP'].mean().reset_index()

df['Cumulative'] = df['VWAP'].cumsum()

plt.figure(figsize=(10, 6))
for i, row in yearly_open_prices.iterrows():
    #plt.bar(row['encoded_year'], row['VWAP'], color='b', label='VWAP' if i == 0 else '')
    plt.bar(row['encoded_year'], df.loc[df['encoded_year'] == row['encoded_year'], 'Cumulative'].iloc[-1],
            color='g' if df['VWAP'].iloc[i] > 0 else 'r', label='VWAP' if i == 0 else '')

plt.xticks(yearly_open_prices['encoded_year'], le.inverse_transform(yearly_open_prices['encoded_year']),rotation='vertical')
plt.title('Waterfall Plot of VWAP')
plt.xlabel('Year')
plt.ylabel('Cumulative Value')
plt.legend()
plt.show()
```

***Conclusion***

The VWAP (Volume-Weighted Average Price) bar plot of a company's stocks is consistently increasing from the year 2000 to 2021, it suggests a generally positive trend in the average price of the stocks over that period. The upward trend in VWAP suggests potential long-term growth in the stock prices and a sign of a healthy and growing company. It could also indicate that investors are optimistic about the company's future earnings and growth potential. It is a positive sign for big return investement.

***Justification***

VWAP stands for "Volume-Weighted Average Price." It is a trading indicator that represents the average price of a security over a specified time period, weighted by the volume of shares traded at each price level. 
It is used as a benchmark to assess whether they achieved better or worse prices for their trades compared to the average market participant. VWAP is sometimes used as a trend indicator. If the current price is above the VWAP, it may be considered a potential upward momentum. Conversely, if the price is below the VWAP, it may be seen as downward momentum.

So, we have plotted a waterfall graph of VWAP to look at the continous trends of the company yearwise.

A waterfall graph, also known as a waterfall chart or bridge chart, serves several purposes in data visualization. This type of chart is particularly useful for illustrating the cumulative effect of sequentially introduced positive or negative values. They are effective for analyzing transitions between different states or values. They provide a clear picture of how each data point contributes to the total and how the total changes over time or categories.

```python
data = df[df['Close'].notnull() & df['Open'].notnull()]
a_close, loc_close, scale_close = skewnorm.fit(data['Close'])

a_open, loc_open, scale_open = skewnorm.fit(data['Open'])

x = np.linspace(min(data['Close'].min(), data['Open'].min()),
                max(data['Close'].max(), data['Open'].max()), 100)

plt.figure(figsize=(10, 6))
plt.hist(data['Close'], bins=30, density=True, alpha=0.6, color='g', label='Close')
plt.hist(data['Open'], bins=30, density=True, alpha=0.3, color='b', label='Open')
plt.plot(x, skewnorm.pdf(x, a_close, loc_close, scale_close), 'k', linewidth=4, label='Skewnorm - Close')
plt.plot(x, skewnorm.pdf(x, a_open, loc_open, scale_open), 'r', linewidth=2, label='Skewnorm - Open')
plt.title('Skewnorm Distribution for Reliance Closing and Opening Prices')
plt.xlabel('Price')
plt.ylabel('Density')
plt.legend()
plt.show()
```
